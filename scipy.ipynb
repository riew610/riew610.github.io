```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
from scipy.optimize import minimize
```

```python
f = lambda x : (x-3)**2
res = minimize(f,1)
res.fun, res.x
```

```python
f = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2
cons = ({'type': 'ineq', 'fun': lambda x:  x[0] - 2 * x[1] + 2},
        {'type': 'ineq', 'fun': lambda x: -x[0] - 2 * x[1] + 6},
        {'type': 'ineq', 'fun': lambda x: -x[0] + 2 * x[1] + 2})
bnds = ((0, None), (0, None))
res = minimize(f, (2, 0), bounds=bnds, constraints=cons)
res.fun, res.x
```

```python
x = np.linspace(0, 10, 10)
y = x**2 * np.sin(x)
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind='cubic')
x_dense = np.linspace(0,10,100)
y_dense = f(x_dense)
plt.scatter(x,y)
plt.plot(x_dense, y_dense)
plt.show()
```

```python
from scipy.special import legendre
x = np.linspace(0,1,100)
plt.plot(x,legendre(6)(x))
plt.show()
```

```python
from scipy.misc import derivative
f = lambda x : x**2 *np.sin(2*x) *np.exp(-x)
x = np.linspace(0,1,100)
plt.plot(x,f(x))
plt.plot(x,derivative(f,x,dx=1e-6))
plt.plot(x,derivative(f,x,dx=1e-6,n=2))
plt.grid()
plt.show()
```

```python
from scipy.integrate import quad
integrand = lambda x : x **2 + np.sin(2*x) * np.exp(-x)
quad(integrand,0,10)
```

```python
from scipy.integrate import dblquad
integrand = lambda y,x : np.sin(x+y**2)
lwr_y = lambda x : -x
upr_y = lambda x : x**2
dblquad(integrand,0,1,lwr_y,upr_y)
```

```python
from scipy.integrate import dblquad
integrand = lambda x,y : np.sin(y+x**2)
lwr_y = lambda x : -x
upr_y = lambda x : x**2
dblquad(integrand,0,1,lwr_y,upr_y)
```

```python
from scipy.integrate import odeint
dvdt = lambda v,t : 3*v**2 -5
v0=0
```

```python
t = np.linspace(0,1,100)
sol = odeint(dvdt,v0,t)
```

```python
plt.plot(t,sol.T[0])
```
