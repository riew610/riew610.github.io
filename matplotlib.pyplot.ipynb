```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

###  $noised \,sin(c) = $ $sin(x_i) + c\epsilon_i$, $\,$ $\epsilon_i \sim N(0,1)$

```python
x = np.linspace(0,20,30)
x1 = np.linspace(0,20,200)
y = np.sin(x) + np.random.randn(len(x))*0.2
plt.figure(figsize=(8,5))
plt.plot(x,y,'*--',color='darkgreen',label = 'noised sin(0.2)')
plt.plot(x,np.sin(x) + np.random.randn(len(x))*0.4,'o--',color = 'brown',label = 'nosied sin(0.4)')
plt.plot(x1,np.sin(x1),'-',color='green',label = 'sin')
plt.grid(True)
plt.xlabel(r'$\tau$',fontsize=15)
plt.ylabel(r'$\lambda$',fontsize=15)
plt.title('sin with noise',fontsize=25,color='brown')
plt.legend(loc= 'upper right', fontsize = 10,ncol=3)
plt.ylim(-2,2)
plt.xlim(-1,21)
plt.show()
```

```python
res = np.random.randn(1000)
plt.figure(figsize = (10,5))
plt.hist(res,bins = 100,density = True,histtype='step')
plt.hist(res,bins = 22, density = True)
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(8,8))
fig.suptitle('trigonometric functions',fontsize=20,color='green')
x = np.linspace(0,1,100)
axes[0][0].plot(x,np.tan(x),'--',color='green',label = r'$\tan(x)$')
axes[0][0].text(0.5,0.1,'standard form')
axes[0][0].set_xlabel('x')
axes[0][0].set_ylabel('y')
axes[0][0].tick_params(axis='both', which='major', labelsize=10)
axes[0][0].legend(loc = 'upper left')
axes[0][0].grid('minor')
axes[0][0].set_title('tan')
#fig.tight_layout()

axes[0][1].plot(x,np.sin(x),'--',color='green',label = r'$\cos(x)$')
axes[0][1].text(0.5,0.1,'standard form')
axes[0][1].set_xlabel('x')
axes[0][1].set_ylabel('y')
axes[0][1].tick_params(axis='both', which='major', labelsize=10)
axes[0][1].legend(loc = 'upper left')
axes[0][1].grid('minor')
axes[0][1].set_title('cos')

axes[1][0].plot(x,np.arctan(x),'--',color='green',label = r'$\arctan(x)$')
axes[1][0].text(0.5,0.1,'inverse form')
axes[1][0].set_xlabel('x')
axes[1][0].set_ylabel('y')
axes[1][0].tick_params(axis='both', which='major', labelsize=10)
axes[1][0].legend(loc = 'upper left')
axes[1][0].grid('minor')
axes[1][0].set_title('arctan')

axes[1][1].plot(x,np.arccos(x),'--',color='green',label = r'$\arccos(x)$')
axes[1][1].text(0.5,0.1,'inverse form')
axes[1][1].set_xlabel('x')
axes[1][1].set_ylabel('y')
axes[1][1].tick_params(axis='both', which='major', labelsize=10)
axes[1][1].legend(loc = 'upper left')
axes[1][1].grid('minor')
axes[1][1].set_title('arccos')

plt.show()
```

```python
ax = plt.axes(projection='3d')
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline,'gray')
zdata = 15*np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata,ydata,zdata,c=zdata,cmap='Reds')
```
