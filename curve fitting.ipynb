```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
x_data = np.array([0.        , 0.15789474, 0.31578947, 0.47368421, 0.63157895,
       0.78947368, 0.94736842, 1.10526316, 1.26315789, 1.42105263,
       1.57894737, 1.73684211, 1.89473684, 2.05263158, 2.21052632,
       2.36842105, 2.52631579, 2.68421053, 2.84210526, 3.        ])
y_data = np.array([  2.95258285,   2.49719803,  -2.1984975 ,  -4.88744346,
        -7.41326345,  -8.44574157, -10.01878504, -13.83743553,
       -12.91548145, -15.41149046, -14.93516299, -13.42514157,
       -14.12110495, -17.6412464 , -16.1275509 , -16.11533771,
       -15.66076021, -13.48938865, -11.33918701, -11.70467566])
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

```python
model_f = lambda x , a , b , c : a*(x-b)**2 + c
```

```python
popt, pcov = curve_fit(model_f, x_data,y_data,p0=[1,1,1])
```

```python
a,b,c = popt
```

```python
x_model = np.linspace(min(x_data),max(x_data),100)
y_model = model_f(x_model,a,b,c)
plt.scatter(x_data,y_data,color='green')
plt.plot(x_model,y_model,color='gray')
plt.show()
```

```python
plt.imshow((np.abs(pcov)))
plt.colorbar()
plt.show()
```

```python
g = lambda x,a,b,c,d : a*(x-b)**2 + c + d*0.0001*np.cos(x)
```

```python
popt, pcov = curve_fit(g,x_data,y_data,p0=[1,2,3,4])
a,b,c,d = popt
```

```python
model_x = np.linspace(min(x_data),max(x_data), 100)
model_y = g(model_x, a,b,c,d)
```

```python
plt.scatter(x_data,y_data)
plt.plot(model_x,model_y,color='gray')
plt.show()
```

```python
plt.imshow((np.log(np.abs(pcov))))
plt.colorbar()
plt.show()
```
